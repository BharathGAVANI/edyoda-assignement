```python
#2.Reverse a linked list in groups of given size
class ListNode:
    def __init__(self, value):
        self.value = value
        self.next = None

def reverseInGroups(head, group_size):
    if not head or group_size <= 1:
        return head

    dummy = ListNode(0)
    dummy.next = head
    prev = dummy
    current = head
    counter = 0

    while current:
        counter += 1
        if counter % group_size == 0:
            prev = reverseGroup(prev, current.next)
            current = prev.next
        else:
            current = current.next

    return dummy.next

def reverseGroup(prev, next_group_start):
    last = prev.next
    current = last.next

    while current != next_group_start:
        last.next = current.next
        current.next = prev.next
        prev.next = current
        current = last.next

    return last

```

```python
#3.Merge a linked list into another linked list at alternate positions.

class ListNode:
    def __init__(self, value):
        self.value = value
        self.next = None

def mergeAtAlternatePositions(main_list, list_to_merge):
    if not main_list or not list_to_merge:
        return main_list

    current_main = main_list
    current_merge = list_to_merge

    while current_main and current_merge:
        # Store next pointers
        next_main = current_main.next
        next_merge = current_merge.next

        # Merge the node from the second list into the main list
        current_main.next = current_merge
        current_merge.next = next_main

        # Move to the next nodes
        current_main = next_main
        current_merge = next_merge

    return main_list
```

```python
#4.In an array, Count Pairs with given sum
def countPairsWithSum(arr, target_sum):
    count = 0  
    seen = {}  

    for num in arr:
        complement = target_sum - num

        if complement in seen:
            count += seen[complement]

        seen[num] = seen.get(num, 0) + 1

    return count // 2  
```

```python
#5.Find duplicates in an array
def findDuplicates(arr):
    seen = set()
    duplicates = []

    for num in arr:
        if num in seen:
            duplicates.append(num)
        else:
            seen.add(num)

    return duplicates

```

```python
#6.Find the Kth largest and Kth smallest number in an array
def findKthLargestAndSmallest(arr, k):
    if k < 1 or k > len(arr):
        return None

    arr.sort()  
    kth_smallest = arr[k - 1]  
    kth_largest = arr[-k]  

    return kth_largest, kth_smallest
```

```python
#7.Move all the negative elements to one side of the array
def moveNegativesToLeft(arr):
    left = 0  
    right = len(arr) - 1  

    while left <= right:
        if arr[left] < 0 and arr[right] < 0:
            left += 1
        elif arr[left] >= 0 and arr[right] < 0:
            arr[left], arr[right] = arr[right], arr[left]
            left += 1
            right -= 1
        elif arr[left] >= 0 and arr[right] >= 0:
            right -= 1
        else:
            left += 1
            right -= 1

    return arr

```

```python
#8.Reverse a string using a stack data structure
def reverseString(input_string):
    stack = []  

    for char in input_string:
        stack.append(char)

    reversed_string = ""

    while len(stack) > 0:
        reversed_string += stack.pop()

    return reversed_string
```

```python
#9.Evaluate a postfix expression using stack
def evaluatePostfix(expression):
    stack = []  

    operators = set("+-*/")  

    for char in expression:
        if char not in operators:
            stack.append(int(char))
        else:
            operand2 = stack.pop()
            operand1 = stack.pop()
            result = performOperation(operand1, operand2, char)
            stack.append(result)

    return stack[0]

def performOperation(operand1, operand2, operator):
    if operator == '+':
        return operand1 + operand2
    elif operator == '-':
        return operand1 - operand2
    elif operator == '*':
        return operand1 * operand2
    elif operator == '/':
        if operand2 == 0:
            raise ValueError("Division by zero")
        return operand1 // operand2  

postfix_expression = "34+5*"
result = evaluatePostfix(postfix_expression)
print("Result:", result)

```

```python
#10.Implement a queue using the stack data structure
class QueueUsingStacks:
    def __init__(self):
        self.enqueue_stack = [] 
        self.dequeue_stack = []  

    def enqueue(self, item):
        self.enqueue_stack.append(item)  

    def dequeue(self):
        if not self.dequeue_stack:
            if not self.enqueue_stack:
                return None
            while self.enqueue_stack:
                self.dequeue_stack.append(self.enqueue_stack.pop())

        return self.dequeue_stack.pop()


queue = QueueUsingStacks()
queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)

print(queue.dequeue())  
print(queue.dequeue())  
queue.enqueue(4)
print(queue.dequeue())  
print(queue.dequeue())  
```
