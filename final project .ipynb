```python
#admin block 
food_items = {}

def generate_food_id():
    if len(food_items) == 0:
        return 1
    else:
        return max(food_items.keys()) + 1

def add_food_item():
    food_id = generate_food_id()
    name = input("Enter the name of the food item: ")
    quantity = input("Enter the quantity of the food item: ")
    price = float(input("Enter the price of the food item: "))
    discount = float(input("Enter the discount of the food item: "))
    stock = int(input("Enter the stock of the food item: "))
    
    food_items[food_id] = {
        'Name': name,
        'Quantity': quantity,
        'Price': price,
        'Discount': discount,
        'Stock': stock
    }
    
    print("Food item added successfully.")

def edit_food_item():
    food_id = int(input("Enter the FoodID of the food item you want to edit: "))
    
    if food_id in food_items:
        print("Current details:")
        print(food_items[food_id])
        
        name = input("Enter the new name of the food item: ")
        quantity = input("Enter the new quantity of the food item: ")
        price = float(input("Enter the new price of the food item: "))
        discount = float(input("Enter the new discount of the food item: "))
        stock = int(input("Enter the new stock of the food item: "))
        
        food_items[food_id]['Name'] = name
        food_items[food_id]['Quantity'] = quantity
        food_items[food_id]['Price'] = price
        food_items[food_id]['Discount'] = discount
        food_items[food_id]['Stock'] = stock
        
        print("Food item updated successfully.")
    else:
        print("Food item not found.")

def view_food_items():
    if len(food_items) == 0:
        print("No food items found.")
    else:
        print("List of food items:")
        for food_id, details in food_items.items():
            print(f"FoodID: {food_id}")
            print(f"Name: {details['Name']}")
            print(f"Quantity: {details['Quantity']}")
            print(f"Price: {details['Price']}")
            print(f"Discount: {details['Discount']}")
            print(f"Stock: {details['Stock']}")
            print("")

def remove_food_item():
    food_id = int(input("Enter the FoodID of the food item you want to remove: "))
    
    if food_id in food_items:
        del food_items[food_id]
        print("Food item removed successfully.")
    else:
        print("Food item not found.")

def main_menu():
    while True:
        print("1. Add new food item")
        print("2. Edit food item")
        print("3. View all food items")
        print("4. Remove food item")
        print("0. Exit")
        
        choice = input("Enter your choice: ")
        
        if choice == "1":
            add_food_item()
        elif choice == "2":
            edit_food_item()
        elif choice == "3":
            view_food_items()
        elif choice == "4":
            remove_food
```

```python
#user id 
users = {}
order_history = []

def register():
    full_name = input("Enter your full name: ")
    phone_number = input("Enter your phone number: ")
    email = input("Enter your email: ")
    address = input("Enter your address: ")
    password = input("Enter your password: ")
    
    user_id = len(users) + 1
    
    users[user_id] = {
        'Full Name': full_name,
        'Phone Number': phone_number,
        'Email': email,
        'Address': address,
        'Password': password
    }
    
    print("Registration successful.")

def login():
    email = input("Enter your email: ")
    password = input("Enter your password: ")
    
    for user_id, user_details in users.items():
        if user_details['Email'] == email and user_details['Password'] == password:
            return user_id
    
    return None

def display_main_menu(user_id):
    while True:
        print("\nMain Menu:")
        print("1. Place New Order")
        print("2. Order History")
        print("3. Update Profile")
        print("0. Logout")
        
        choice = input("Enter your choice: ")
        
        if choice == "1":
            place_new_order(user_id)
        elif choice == "2":
            view_order_history(user_id)
        elif choice == "3":
            update_profile(user_id)
        elif choice == "0":
            print("Logged out.")
            break

def display_food_menu():
    print("\nFood Menu:")
    print("1. Tandoori Chicken (4 pieces) [INR 240]")
    print("2. Vegan Burger (1 Piece) [INR 320]")
    print("3. Truffle Cake (500gm) [INR 900]")

def place_new_order(user_id):
    display_food_menu()
    
    selected_items = []
    while True:
        try:
            item_numbers = input("Enter the numbers corresponding to the food items (comma-separated): ")
            item_numbers = [int(num) for num in item_numbers.split(",")]
            if not all(1 <= num <= 3 for num in item_numbers):
                raise ValueError
            break
        except ValueError:
            print("Invalid input. Please enter valid numbers.")
    
    for item_number in item_numbers:
        if item_number == 1:
            selected_items.append("Tandoori Chicken")
        elif item_number == 2:
            selected_items.append("Vegan Burger")
        elif item_number == 3:
            selected_items.append("Truffle Cake")
    
    print("\nSelected items:")
    for item in selected_items:
        print(item)
    
    place_order = input("\nDo you want to place the order? (yes/no): ")
    if place_order.lower() == "yes":
        total_amount = 0
        for item_number in item_numbers:
            if item_number == 1:
                total_amount += 240
            elif item_number == 2:
                total_amount += 320
            elif item_number == 3:
                total_amount += 900
        
        order_history.append({
            'User ID': user_id,
            'Items': selected_items,
            'Total Amount': total_amount
        })
        
        print("Order placed successfully.")
    else:
        print("Order canceled.")

def view_order_history(user_id):
    print("\nOrder History:")
    user_orders = [order for order in order_history if order['User ID'] == user_id]
    
    if len(user_orders) == 0:
        print("No order history found.")
    else:
        for order in user_orders:
            print("Items:", ', '.join(order['Items']))
            print("Total Amount: INR", order['Total Amount'])
            print("--------------------")
    
def update_profile(user_id):
    print("\nUpdate Profile:")
    
    if user_id in users:
        user_details = users[user_id]
        print("Current details:")
        print("Full Name:", user_details['Full Name'])
        print("Phone Number:", user_details['Phone Number'])
        print("Email:", user_details['Email'])
        print("Address:", user_details['Address'])
        
        full_name = input("Enter your new full name: ")
        phone_number = input("Enter your new phone number: ")
        email = input("Enter your new email: ")
        address = input("Enter your new address: ")
        
        users[user_id]['Full Name'] = full_name
        users[user_id]['Phone Number'] = phone_number
        users[user_id]['Email'] = email
        users[user_id]['Address'] = address
        
        print("Profile updated successfully.")
    else:
        print("User not found.")

def main_menu():
    while True:
        print("\n1. Register")
        print("2. Log in")
        print("0. Exit")
        
        choice = input("Enter your choice: ")
        
        if choice == "1":
            register()
        elif choice == "2":
            user_id = login()
            if user_id is not None:
                print("Login successful.")
                display_main_menu(user_id)
            else:
                print("Invalid email or password.")
        elif choice == "0":
            print("Exiting the application.")
            break
        else:
            print("Invalid choice.")

main_menu()
```
